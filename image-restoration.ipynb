```python
class Gopro(data.Dataset):
    def __init__(self, data_dir, window_size=128, is_train=False):
        super(Gopro, self).__init__()
        self.is_train = is_train
        self.window_size = window_size

        self.ground_truth_file_names = []

        for directory_name in os.listdir(data_dir):
            ground_truth_images = os.path.join(data_dir, directory_name, 'sharp')
            groundtruth_image_file_names = os.listdir(ground_truth_images)

            for path in groundtruth_image_file_names:
                ground_truth_file_name = os.path.join(ground_truth_images, path)
                self.ground_truth_file_names.append(ground_truth_file_name)

        self.n_samples = len(self.ground_truth_file_names)

    def __getitem__(self, idx):

        sharp_file_path = self.ground_truth_file_names[idx]
        blur_file_path = sharp_file_path.replace("sharp", "blur")

        train_image = Image.open(blur_file_path).convert('RGB')
        groundtruth_image = Image.open(sharp_file_path).convert('RGB')

        # if training is on, augement the dataset with affine and color channel transformations 
        if self.is_train:
              w, h = train_image.size
              p = self.window_size
              x = random.randrange(0, w - p + 1)
              y = random.randrange(0, h - p + 1)
              train_image = train_image.crop((x, y, x + p, y + p))
              groundtruth_image = groundtruth_image.crop((x, y, x + p, y + p))

              degree = random.choice([0, 90, 180, 270])
              train_image = transforms.functional.rotate(train_image, degree)
              groundtruth_image = transforms.functional.rotate(groundtruth_image, degree)

    
              train_image = transforms.functional.adjust_gamma(train_image, 1)
              groundtruth_image = transforms.functional.adjust_gamma(groundtruth_image, 1)
              sat_factor = 1 + (0.2 - 0.4 * np.random.rand())
              train_image = transforms.functional.adjust_saturation(train_image, sat_factor)
              groundtruth_image = transforms.functional.adjust_saturation(groundtruth_image, sat_factor)
            
        input_b1 = transforms.ToTensor()(train_image)
        target_s1 = transforms.ToTensor()(groundtruth_image)

        H = input_b1.size()[1]
        W = input_b1.size()[2]
        
        return {'input_b1': input_b1, 'target_s1': target_s1,'blurfile_path':blur_file_path}

    def __len__(self):
        return self.n_samples
```

# Utils - functions to calculate metrics and checkpointing of model

```python
def tensor_to_rgb(img_input):
    output = img_input.cpu()
    output = output.data.squeeze(0)

    output = output.numpy()
    output *= 255.0
    output = output.clip(0, 255)

    return output


def calculate_psnr(img1, img2):
    mse = ((img1 - img2) ** 2).mean()
    psnr = 10 * log10(255 * 255 / (mse + 10 ** (-10)))
    return psnr


class SaveData():
    def __init__(self, save_dir, exp_name, finetuning):
        self.save_dir = os.path.join(save_dir, exp_name)

        if not finetuning:
            if os.path.exists(self.save_dir):
                os.system('rm -rf ' + self.save_dir)
            #os.makedirs(self.save_dir)
        else:
            if not os.path.exists(self.save_dir):
                os.makedirs(self.save_dir)

        self.save_dir_model = os.path.join(self.save_dir, 'model')
        if not os.path.exists(self.save_dir_model):
            os.makedirs(self.save_dir_model)

        self.logFile = open(self.save_dir + '/log.txt', 'a')

        save_dir_tensorboard = os.path.join(self.save_dir, 'logs')
        if not os.path.exists(save_dir_tensorboard):
            os.makedirs(save_dir_tensorboard)
        self.writer = SummaryWriter(save_dir_tensorboard)


    def save_params(self, args):
        with open(self.save_dir + '/params.txt', 'w') as params_file:
            params_file.write(str(args.__dict__) + "\n\n")


    def save_model(self, model, epoch,optimizer):
        torch.save(model.state_dict(), self.save_dir_model + '/model_lastest.pt')
        torch.save(model.state_dict(), self.save_dir_model + '/model_lastest_copy.pt')
        torch.save(optimizer.state_dict(), self.save_dir_model + '/optimizer_lastest.pt')

        torch.save(epoch, self.save_dir_model + '/last_epoch.pt')

    def save_log(self, log):
        sys.stdout.flush()
        self.logFile.write(log + '\n')
        self.logFile.flush()

    def load_model(self, model,optimizer=None):
        model.load_state_dict(torch.load(self.save_dir_model + '/model_lastest.pt'))

        if optimizer is not None:
            optimizer.load_state_dict(torch.load(self.save_dir_model + '/optimizer_lastest.pt'))

        last_epoch = torch.load(self.save_dir_model + '/last_epoch.pt')

        print("Load mode_status from {}/model_lastest.pt, epoch: {}".format(self.save_dir_model, last_epoch))
        return model, last_epoch,optimizer

    def add_scalar(self, tag, value, step):
        self.writer.add_scalar(tag, value, step)

# Calculate Gram matrix (G = FF^T)
def gram(x):
    (bs, ch, h, w) = x.size()
    f = x.view(bs, ch, w*h)
    f_T = f.transpose(1, 2)
    G = f.bmm(f_T) / (ch * h * w)
    return G

def gaussian(window_size, sigma):
    gauss = torch.Tensor([exp(-(x - window_size // 2) ** 2 / float(2 * sigma ** 2)) for x in range(window_size)])
    return gauss / gauss.sum()


def create_window(window_size, channel):
    _1D_window = gaussian(window_size, 1.5).unsqueeze(1)
    _2D_window = _1D_window.mm(_1D_window.t()).float().unsqueeze(0).unsqueeze(0)
    window = Variable(_2D_window.expand(channel, 1, window_size, window_size).contiguous())
    return window


def _ssim(img1, img2, window, window_size, channel, size_average=True):
    mu1 = F.conv2d(img1, window, padding=window_size // 2, groups=channel)
    mu2 = F.conv2d(img2, window, padding=window_size // 2, groups=channel)

    mu1_sq = mu1.pow(2)
    mu2_sq = mu2.pow(2)
    mu1_mu2 = mu1 * mu2

    sigma1_sq = F.conv2d(img1 * img1, window, padding=window_size // 2, groups=channel) - mu1_sq
    sigma2_sq = F.conv2d(img2 * img2, window, padding=window_size // 2, groups=channel) - mu2_sq
    sigma12 = F.conv2d(img1 * img2, window, padding=window_size // 2, groups=channel) - mu1_mu2

    C1 = 0.01 ** 2
    C2 = 0.03 ** 2

    ssim_map = ((2 * mu1_mu2 + C1) * (2 * sigma12 + C2)) / ((mu1_sq + mu2_sq + C1) * (sigma1_sq + sigma2_sq + C2))

    if size_average:
        return ssim_map.mean()
    else:
        return ssim_map.mean(1).mean(1).mean(1)


class SSIM(torch.nn.Module):
    def __init__(self, window_size=11, size_average=True):
        super(SSIM, self).__init__()
        self.window_size = window_size
        self.size_average = size_average
        self.channel = 1
        self.window = create_window(window_size, self.channel)

    def forward(self, img1, img2):
        (_, channel, _, _) = img1.size()

        if channel == self.channel and self.window.data.type() == img1.data.type():
            window = self.window
        else:
            window = create_window(self.window_size, channel)

            if img1.is_cuda:
                window = window.cuda(img1.get_device())
            window = window.type_as(img1)

            self.window = window
            self.channel = channel

        return _ssim(img1, img2, window, self.window_size, channel, self.size_average)


def ssim(img1, img2, window_size=11, size_average=True):
    (_, channel, _, _) = img1.size()
    window = create_window(window_size, channel)

    if img1.is_cuda:
        window = window.cuda(img1.get_device())
    window = window.type_as(img1)

    return _ssim(img1, img2, window, window_size, channel, size_average)
```

# Deformable Convolution block

```python
#REF PAPER: https://arxiv.org/pdf/1811.11168.pdf
#REF https://github.com/oeway/pytorch-deform-conv
class DeformableConv2d_Custom(nn.Module):
    def __init__(self,channels_input,out_channels,
                 conv_kernel_size=3,stride=1,padding=1,bias=False):

        super(DeformableConv2d_Custom, self).__init__()
        # conv_kernel_size = (conv_kernel_size, conv_kernel_size)
        self.stride = (stride, stride)
        self.padding = padding
        self.offset_conv = nn.Conv2d(channels_input,2 * conv_kernel_size* conv_kernel_size,
                                     kernel_size=conv_kernel_size,stride=stride,
                                     padding=self.padding,bias=True)

        nn.init.constant_(self.offset_conv.weight, 0.)
        nn.init.constant_(self.offset_conv.bias, 0.)
        
        self.v2_modulation_conv2d = nn.Conv2d(channels_input, 1 * conv_kernel_size * conv_kernel_size,
                                     kernel_size=conv_kernel_size,stride=stride,padding=self.padding,bias=True)

        nn.init.constant_(self.v2_modulation_conv2d.weight, 0.)
        nn.init.constant_(self.v2_modulation_conv2d.bias, 0.)
        
        self.conv_layer_1 = nn.Conv2d(in_channels=channels_input,out_channels=out_channels,kernel_size=conv_kernel_size,stride=stride,
                                      padding=self.padding,bias=bias)

    def forward(self, x):
        
        offset = self.offset_conv(x)
        conv_output = self.v2_modulation_conv2d(x)
        mask = 2. * torch.sigmoid(conv_output)
        
        x = torchvision.ops.deform_conv2d(input=x,offset=offset,weight=self.conv_layer_1.weight,bias=self.conv_layer_1.bias, 
                                          padding=self.padding,mask=mask,stride=self.stride,)
        return x
```

# Attention modules

```python
#REF. https://arxiv.org/pdf/2202.09741.pdf
class LargeKernelAttention(nn.Module):
  def __init__(self,dim) -> None:
      super().__init__()
      self.step1 = nn.Conv2d(dim,dim,5,padding=2,groups=dim)
      self.step2_spatial_conv = nn.Conv2d(dim,dim,7,stride=1,padding=9,groups=dim,dilation=3)
      self.step3 = nn.Conv2d(dim,dim,1)

  def forward(self,x):
    residual = x.clone()
    out1 = self.step1(x)
    out1 = self.step2_spatial_conv(x)
    out1 = self.step3(x)
    return residual * out1
    
class ChannelAttention(nn.Module):
    def __init__(self, input_channels):
      super(ChannelAttention, self).__init__()
      self.global_avg_pooling = nn.AdaptiveAvgPool2d(1)
      self.multi_layer_perceptron = nn.Sequential(
            nn.Linear(input_channels, input_channels // 16, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(input_channels // 16, input_channels, bias=False),
            nn.Sigmoid())
      
    def forward(self,x):
        b, c, _, _ = x.size()
        y = self.global_avg_pooling(x).view(b, c)
        y = self.multi_layer_perceptron(y).view(b, c, 1, 1)
        return x * y.expand_as(x)

class ResnetAdaptiveBlock(nn.Module):
    expansion = 1

    def __init__(self, input_channels, output_channels, stride=1, downsample=None):
        super(ResnetAdaptiveBlock, self).__init__()
        self.conv_layer_1 = nn.Conv2d(input_channels, output_channels, kernel_size=3, stride=stride,padding=1, bias=False)
        self.batch_norm_layer_1 = nn.BatchNorm2d(output_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv_layer_2 = nn.Conv2d(input_channels, output_channels, kernel_size=3, stride=stride,padding=1, bias=False)
        self.batch_norm_layer_2 = nn.BatchNorm2d(output_channels)
        
        self.sa = LargeKernelAttention(output_channels)

        self.downsample = downsample
        self.stride = stride
        temp1 = [nn.Conv2d(output_channels, output_channels, kernel_size=3, stride=stride,padding=1, bias=False),
                                      nn.Sigmoid()]
        self.sequential_1 = nn.Sequential(*temp1)
        self.sequential_2 = nn.Sequential(*temp1)
        self.FC = nn.Sequential(nn.Linear(output_channels, output_channels, bias=False),nn.ReLU(True),nn.Linear(output_channels, output_channels, bias=False),
                                nn.ReLU(True))          

    def forward(self, x):
        residual = x

        out = self.conv_layer_1(x)
        out = self.batch_norm_layer_1(out)
        out = self.relu(out)

        out = self.conv_layer_2(out)
        out = self.batch_norm_layer_2(out)

        out_s = self.sa(out) * out
      
        wight1 = self.sequential_1(out_s) 
        out_c1 = out_s*wight1
        wight2 = self.sequential_2(out) 
        out_s2 = out_s*wight2
        
        wight3 = torch.cat([wight1.view(wight1.shape[0],-1).unsqueeze(0), wight1.view(wight1.shape[0],-1).unsqueeze(0)], 0)
        wight3_temp = torch.max(wight3,0)[0]
    
        wight3 = 1-wight3_temp.view(wight1.shape)
        out_temp = out_s*wight3
        out_s3 = self.sa(out_temp) * out_temp
        out = out_c1 + out_s2 + out_s3

        if self.downsample is not None:
            residual = self.downsample(x)

        out += residual
        out = self.relu(out)

        return out
```

# ResNet implementation

```python
#REFERENCE CODE: https://github.com/pytorch/vision/blob/main/torchvision/models/resnet.py
class ResnetModel(nn.Module):
    def __init__(self, input_channels, output_channels, features=64, resnet_blocks=6, adaNorm = False,skip=True,deformable=True,dcn_ksize=1):
        assert(resnet_blocks >= 0)
        super(ResnetModel, self).__init__()
        self.input_channels = input_channels
        self.output_channels = output_channels
        self.features = features
        self.resnet_blocks = resnet_blocks
        self.adaNorm = adaNorm
        self.skip = skip
        self.deformable = deformable 
        print("features ::::: ",features)
        intermidiate_block = []
        
        intermidiate_block += [DeformableConv2d_Custom(input_channels,features,dcn_ksize,stride=1, padding=dcn_ksize//2),
                               nn.InstanceNorm2d(features),
                              nn.ReLU(True)]

        intermidiate_block1 = []
        n_downsampling = 2
        for i in range(n_downsampling):
            scale_factor = 2**i
            intermidiate_block1 += [nn.ReflectionPad2d(1),
                          nn.Conv2d(features * scale_factor, features * scale_factor * 2, kernel_size=3, stride=2, padding=0, bias=False),
                          nn.InstanceNorm2d(features * scale_factor * 2),
                          nn.ReLU(True)]
        self.intermidiate_block = nn.Sequential(*intermidiate_block)
        self.intermidiate_block1_1 = nn.Sequential(*intermidiate_block1[0:4])
        self.intermidiate_block1_2 = nn.Sequential(*intermidiate_block1[4:-1])

     
        intermidiate_block2 = []
        scale_factor = 2**n_downsampling
        for i in range(resnet_blocks):
            intermidiate_block2 += [ResnetBlock(features * scale_factor, use_bias=False)]
            
        self.intermidiate_block2 = nn.Sequential(*intermidiate_block2)

        # Class Activation Map
        self.gap_fc = nn.Linear(features * scale_factor, 1, bias=False)
        self.gmp_fc = nn.Linear(features * scale_factor, 1, bias=False)
        self.conv1x1 = nn.Conv2d(features * scale_factor * 2, features * scale_factor, kernel_size=1, stride=1, bias=True)
        self.relu = nn.ReLU(True)

        FC = [nn.Linear(features * scale_factor, features * scale_factor, bias=False),
              nn.ReLU(True),
              nn.Linear(features * scale_factor, features * scale_factor, bias=False),
              nn.ReLU(True)]

        self.gamma = nn.Linear(features * scale_factor, features * scale_factor, bias=False)
        self.beta = nn.Linear(features * scale_factor, features * scale_factor, bias=False)


        for i in range(resnet_blocks):
            setattr(self, 'intermidiate_block_up_' + str(i+1), ResnetExtendedBlock(features * scale_factor, use_bias=False))


        intermidiate_block_up = []
        for i in range(n_downsampling):
            scale_factor = 2**(n_downsampling - i)
            intermidiate_block_up += [nn.Upsample(scale_factor=2, mode='nearest'),
                        nn.Conv2d(2*features * scale_factor, int(features * scale_factor / 2), kernel_size=3, stride=1, padding=1, bias=False),
                        InputLayerNormalization(int(features * scale_factor / 2)),
                        nn.ReLU(True)]
        self.intermidiate_block_up_1 = nn.Sequential(*intermidiate_block_up[0:4])
        self.intermidiate_block_up_2 = nn.Sequential(*intermidiate_block_up[4:-1])
                            
       
        UpBlock_final = []
        UpBlock_final += [DeformableConv2d_Custom(features,output_channels,dcn_ksize,stride=1, padding=dcn_ksize//2), nn.Tanh()]
       

        self.intermidiate_block_up_final = nn.Sequential(*UpBlock_final)


        self.FC = nn.Sequential(*FC)
        self.ResnetAdaptiveBlock = ResnetAdaptiveBlock(features * scale_factor *2, features * scale_factor*2)


    def forward(self, input):
        x = self.intermidiate_block(input)
        x_d1 = self.intermidiate_block1_1(x)
        x_d2 = self.intermidiate_block1_2(x_d1)
        x = self.intermidiate_block2(x_d2)
        x = self.ResnetAdaptiveBlock(x) 
        if self.adaNorm:
            x_ = torch.nn.functional.adaptive_avg_pool2d(x, 1)
            x_ = self.FC(x_.view(x_.shape[0], -1))
            gamma, beta = self.gamma(x_), self.beta(x_) # 1*128 map 1*128

            for i in range(self.resnet_blocks):
                x = getattr(self, 'intermidiate_block_up_' + str(i+1))(x, gamma, beta)
        else:
            x = self.intermidiate_block2(x)

        x = torch.cat([x,x_d2],1)
        x = self.intermidiate_block_up_1(x)

        x = torch.cat([x, x_d1], 1)
        x = self.intermidiate_block_up_2(x)

        out = self.intermidiate_block_up_final(x)
        
        
        if self.skip:
            out = out + input   
        return out     


class ResnetBlock(nn.Module):
    def __init__(self, dim, use_bias):
        super(ResnetBlock, self).__init__()
        conv_block = []
        conv_block += [nn.ReflectionPad2d(1),
                       nn.Conv2d(dim, dim, kernel_size=3, stride=1, padding=0, bias=use_bias),
                       nn.InstanceNorm2d(dim),
                       nn.ReLU(True)]

        conv_block += [nn.ReflectionPad2d(1),
                       nn.Conv2d(dim, dim, kernel_size=3, stride=1, padding=0, bias=use_bias),
                       nn.InstanceNorm2d(dim)]

        self.conv_block = nn.Sequential(*conv_block)

    def forward(self, x):
        out = x + self.conv_block(x)
        return out


class ResnetExtendedBlock(nn.Module):
    def __init__(self, dim, use_bias):
        super(ResnetExtendedBlock, self).__init__()
        self.pad1 = nn.ReflectionPad2d(1)
        self.conv1 = nn.Conv2d(dim, dim, kernel_size=3, stride=1, padding=0, bias=use_bias)
        self.norm1 = AdaptiveInputNormalization(dim)
        self.relu1 = nn.ReLU(True)

        self.pad2 = nn.ReflectionPad2d(1)
        self.conv2 = nn.Conv2d(dim, dim, kernel_size=3, stride=1, padding=0, bias=use_bias)
        self.norm2 = AdaptiveInputNormalization(dim)

    def forward(self, x, gamma, beta):
        out = self.pad1(x)
        out = self.conv1(out)
        out = self.norm1(out, gamma, beta)
        out = self.relu1(out)
        out = self.pad2(out)
        out = self.conv2(out)
        out = self.norm2(out, gamma, beta)

        return out + x


class AdaptiveInputNormalization(nn.Module):
    def __init__(self, num_features, eps=1e-5):
        super(AdaptiveInputNormalization, self).__init__()
        self.eps = eps
        self.rho = Parameter(torch.Tensor(1, num_features, 1, 1))
        self.rho.data.fill_(0.9)

    def forward(self, input, gamma, beta):
        in_mean, in_var = torch.mean(input, dim=[2, 3], keepdim=True), torch.var(input, dim=[2, 3], keepdim=True)
        out_in = (input - in_mean) / torch.sqrt(in_var + self.eps)
        ln_mean, ln_var = torch.mean(input, dim=[1, 2, 3], keepdim=True), torch.var(input, dim=[1, 2, 3], keepdim=True)
        out_ln = (input - ln_mean) / torch.sqrt(ln_var + self.eps)
        out = self.rho.expand(input.shape[0], -1, -1, -1) * out_in + (1-self.rho.expand(input.shape[0], -1, -1, -1)) * out_ln
        out = out * gamma.unsqueeze(2).unsqueeze(3) + beta.unsqueeze(2).unsqueeze(3)

        return out


class InputLayerNormalization(nn.Module):
    def __init__(self, num_features, eps=1e-5):
        super(InputLayerNormalization, self).__init__()
        self.eps = eps
        self.rho = Parameter(torch.Tensor(1, num_features, 1, 1))
        self.gamma = Parameter(torch.Tensor(1, num_features, 1, 1))
        self.beta = Parameter(torch.Tensor(1, num_features, 1, 1))
        self.rho.data.fill_(0.0)
        self.gamma.data.fill_(1.0)
        self.beta.data.fill_(0.0)

    def forward(self, input):
        in_mean, in_var = torch.mean(input, dim=[2, 3], keepdim=True), torch.var(input, dim=[2, 3], keepdim=True)
        out_in = (input - in_mean) / torch.sqrt(in_var + self.eps)
        ln_mean, ln_var = torch.mean(input, dim=[1, 2, 3], keepdim=True), torch.var(input, dim=[1, 2, 3], keepdim=True)
        out_ln = (input - ln_mean) / torch.sqrt(ln_var + self.eps)
        out = self.rho.expand(input.shape[0], -1, -1, -1) * out_in + (1-self.rho.expand(input.shape[0], -1, -1, -1)) * out_ln
        out = out * self.gamma.expand(input.shape[0], -1, -1, -1) + self.beta.expand(input.shape[0], -1, -1, -1)

        return out
```

```python
class Args:
  def __init__(self):
    self.gpu = 0
    self.n_feats = 32
    self.n_resblocks = 9
    self.adaNorm = False 
    self.deformable = True
    self.deformable_kersize = 1 
    self.lr = 1e-4
    self.save_dir = "/content/drive/MyDrive/weights"#"./result" 
    self.data_dir =  "/content/gopro/train"#"/content/SIDD/SIDD_Small_sRGB_Only/"
    self.val_data_dir = "/content/gopro/test"#"/content/SIDD/SIDD_Small_sRGB_Only/"#
    self.exp_name = "CSE676"
    self.finetuning = False
    self.lr_step_size =  1000
    self.lr_gamma = 0.01
    
    self.window_size = 128
    self.batch_size = 1
    self.n_threads = 2
    self.multi = False
    
    self.epochs = 50

    self.period = 1
    
train_args = Args()
```

```python
ssimfunc = ssim
#BUG FIX not running on selected GPU https://stackoverflow.com/questions/37893755/tensorflow-set-cuda-visible-devices-within-jupyter
if train_args.gpu == 0:
    os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
    os.environ["CUDA_VISIBLE_DEVICES"] = "0"
elif train_args.gpu == 1:
    os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
    os.environ["CUDA_VISIBLE_DEVICES"] = "1"
else:
    os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
    os.environ["CUDA_VISIBLE_DEVICES"] = str(train_args.gpu)

loss_values = []
ssim_list = []
psnr_list = []


def get_dataset(data_dir, window_size=None, batch_size=1, n_threads=8, is_train=False, dataset="gopro"):
    
    if dataset == "gopro":
      dataset = Gopro(data_dir, window_size=window_size, is_train=is_train)
    # (self, data_dir, patch_size=128, is_train=False)
    else:
      print("loading sidd")
      dataset = SIDD(data_dir,window_size=window_size,is_train=is_train)
    
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size,
                                             drop_last=True, shuffle=is_train, num_workers=int(n_threads))
    return dataloader

 
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


def validation(model, dataloader, multi):
    total_psnr = 0
    total_ssim = 0
    for batch, images in tqdm(enumerate(dataloader)):
        with torch.no_grad():
            train_image = Variable(images['input_b1'].cuda())
            groundtruth_image = Variable(images['target_s1'].cuda())

            model_output = model(train_image)

        model_output = model_output.cpu()
        model_output = model_output.data.squeeze(0)

        model_output = model_output.numpy()
        model_output *= 255.0
        model_output = model_output.clip(0, 255)
            
        groundtruth_image = groundtruth_image.cpu()
        groundtruth_image = groundtruth_image.data.squeeze(0)

        groundtruth_image = groundtruth_image.numpy()
        groundtruth_image *= 255.0
        groundtruth_image = groundtruth_image.clip(0, 255)


        mean_squared_loss = ((groundtruth_image - model_output) ** 2).mean()
        psnr = 10 * log10(255 * 255 / (mean_squared_loss + 10 ** (-10)))
      
        total_psnr += psnr
        psnr_list.append(psnr)
        
        input_b1  = tensor_to_rgb(train_image)
        

        model_output = model_output.squeeze()
        groundtruth_image = groundtruth_image.squeeze()

        with torch.no_grad():
            ssim = ssimfunc(torch.from_numpy(model_output[None] / 255.0).cuda(),
                                torch.from_numpy(groundtruth_image[None] / 255.0).cuda()).cpu().numpy()
        ssim_list.append(ssim)
        total_ssim += ssim

    return total_psnr / (batch + 1), total_ssim


def train(args):
    print(args)

    model = ResnetModel(3,3,features=args.n_feats,resnet_blocks=args.n_resblocks,
                                        adaNorm=args.adaNorm,deformable=args.deformable,dcn_ksize=args.deformable_kersize)
    model = model.cuda()
   
    loss_function = nn.MSELoss().cuda()
    optimizer = optim.Adam(model.parameters(), lr=args.lr)
    

    # utility for saving models, parameters and logs
    save = SaveData(args.save_dir, args.exp_name, args.finetuning)
    save.save_params(args)
    num_params = count_parameters(model)
    save.save_log(str(num_params))

   
   
    # load pre-trained model if provided
    last_epoch = -1
    if args.finetuning:
        model, last_epoch, optimizer = save.load_model(model,optimizer)
        #optimizer.param_groups[0]['lr']=1e-5
    start_epoch = last_epoch + 1

    

    scheduler = lr_scheduler.StepLR(optimizer, args.lr_step_size, args.lr_gamma,last_epoch=last_epoch)

    # load dataset
    data_loader = get_dataset(args.data_dir, window_size=args.window_size, batch_size=args.batch_size,
                              n_threads=args.n_threads, is_train=True, dataset="gopro")
    if args.val_data_dir:
        valid_data_loader = get_dataset(args.val_data_dir, n_threads=args.n_threads,dataset="gopro")

    
    for epoch in range(start_epoch, args.epochs):
        print("* Epoch {}/{}".format(epoch + 1, args.epochs))
        learning_rate = optimizer.param_groups[0]['lr']
        total_loss = 0
        total_ssim = 0
        for batch, images in tqdm(enumerate(data_loader)):
            input_b1 = Variable(images['input_b1'].cuda())
            # print(images['input_b1'].shape)
            target_s1 = Variable(images['target_s1'].cuda())
            # print(images['target_s1'].shape)
            output_l1 = model(input_b1)
            loss = loss_function(output_l1, target_s1)
            model.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.data.cpu().numpy()
            loss_values.append(loss.data.cpu().numpy())

        loss = total_loss / (batch + 1)
        save.add_scalar('train/loss', loss, epoch)
        
        
        if  epoch % args.period == 0:

            if args.val_data_dir:
                torch.cuda.empty_cache()
                model.eval()
                psnr, total_ssim = validation(model, valid_data_loader, args.multi)
                model.train()

                log = "Epoch {}/{} \t Learning rate: {:.10f} \t Train total_loss: {:.5f} \t * Val PSNR: {:.2f} \t * Val SSIM: {:.2f}".format(
                    epoch + 1, args.epochs, learning_rate, loss, psnr,total_ssim)
                print(log)
                save.save_log(log)
                save.add_scalar('valid/psnr', psnr, epoch)
        else:
            log = "Epoch {}/{} \t Learning rate: {:.10f} \t Train total_loss: {:.5f}\n".format(epoch + 1, args.epochs,
                                                                                              learning_rate, loss)
            print(log)
            save.save_log(log)

        save.save_model(model, epoch,optimizer)
        scheduler.step()




```
